```python
import matplotlib.pyplot as plt
import numpy as np

#yはただの変数、折れ線
y0 = [1, 2 , -5, 2]


#sin関数
x = np.linspace(0, 10, 400)
y = 2 * np.sin(2 * x) +2

#ヒストグラム
x_hist = np.random.normal(15, 5, 2000)

#散布図
np.random.seed()
x_scatter = 4 + np.random.normal(0, 2, 60)
y_scatter = 4 + np.random.normal(0, 2, len(x_scatter))



#2行3列のグラフを作成
fig, axes = plt.subplots(2, 3, tight_layout=True)


axes[0, 0].plot(y0)   #折れ線
axes[0, 1].plot(x, y)   #sin
axes[0, 1].set_xlim(0, 10)
axes[0, 1].set_ylim(-1, 5)
axes[0, 1].set_xticks(range(10))
axes[0, 1].set_yticks(range(-1, 6))

axes[0, 2].hist(x_hist, bins=10)   #ヒストグラム
axes[1, 0].scatter(x_scatter, y_scatter)   #散布図
axes[1, 1]
axes[1, 2]

plt.show()
```
