```python
import pandas as pd
import matplotlib.pyplot as plt
import japanize_matplotlib

# データ読み込み
df = pd.read_csv("/sample_pandas_6.csv")
category_df = pd.read_csv("/category.csv")

df = pd.merge(df, category_df[['商品番号', 'カテゴリー']], how='inner', on='商品番号')

# カテゴリー出現頻度を棒グラフ化
cat_counts = df['カテゴリー'].value_counts()
ax = cat_counts.plot(kind='bar', rot=0)

ax.set_title('カテゴリー別 出現頻度')
ax.set_xlabel('カテゴリー')
ax.set_ylabel('件数')
plt.tight_layout()
plt.show()

# --- 商品番号ごとの注文数の基本統計量 ---
order_stats = df.groupby('商品番号')['注文数'].describe()

print('▼ 商品番号ごとの注文数 基本統計量（先頭5件）')
print(order_stats.head())
```
